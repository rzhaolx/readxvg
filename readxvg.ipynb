```python
# Packages needed
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('figure', figsize = (7,7))
rc('font',**{'size': 16})
rc('text', usetex = False)
rc('mathtext',**{'fontset':'stix'})

# Initialize labels and colours - could be changed as needed
label = ['20gly', 'alt', 'ran', 'dib', '2ran', 'SeqA', 'SeqB', 'SeqC']
colour = ['orangered', 'chocolate', 'orange', 
          'limegreen', 'cyan', 'royalblue', 'violet', 'purple']

# Default window size for running average
window_size = 100

# Default running average function
def get_RunningAvg(data, size):
    '''
    Function that computes the running average of an array
    data: the input array
    size: the window size for computing running average
    WARNING: size must be an even number'''
    i = 100
    res = []
    while i < len(data) - size + 1:
        window = data[i : i + size]
        window_avg = round(np.sum(window) / size, 2)
        res.append(window_avg)
        i += 1
    return res

# Initialize dictionary for results
dict = {}
for i in range(len(label)):
    dict[label[i]] = {}
```

```python
# check if random 2 has a equilibration problem
filelist = ['gyrate_20gly.xvg', 'gyrate_alt.xvg', 'gyrate_ran.xvg', 'gyrate_dib.xvg',
            'gyrate_2ran.xvg', 'gyrate_A.xvg', 'gyrate_B.xvg', 'gyrate_C.xvg']

for i in range(len(filelist)):
    dict[label[i]]['rg'] = {}
    res = np.loadtxt(filelist[i], comments = ['#', '@'], unpack = 'true')
    if label[i] == 'dib':
        dict[label[i]]['rg']['ts'] = res[0][9000::]
        dict[label[i]]['rg']['rs'] = res[1][9000::]
    else:
        dict[label[i]]['rg']['ts'] = res[0][10000::]
        dict[label[i]]['rg']['rs'] = res[1][10000::]
    print(len(dict[label[i]]['rg']['ts']), len(dict[label[i]]['rg']['rs']))

plt.figure(figsize = [10, 7])
plt.plot(dict['2ran']['rg']['ts'], dict['2ran']['rg']['rs'])
plt.xlabel('$R_g$ [nm]')
plt.ylabel('Probability [%]')
#plt.xlim(.48, .92)
plt.legend(loc = 'upper right')
plt.show()
```

```python
### Hydrogen Bonds Within ###

filelist = ['hbnum-int-20gly.xvg', 'hbnum-int-alt.xvg', 'hbnum-int-ran.xvg', 'hbnum-int-dib.xvg', 
            'hbnum-int-2ran.xvg', 'hbnum-int-A.xvg', 'hbnum-int-B.xvg', 'hbnum-int-C.xvg']

# Loop without running average - use this!
for i in range(len(filelist)):
    dict[label[i]]['HBond_int'] = {}
    HBond = np.loadtxt(filelist[i], comments = ['#', '@'], unpack = 'true')
    if label[i] == '20gly':
        dict[label[i]]['HBond_int']['ts'] = HBond[0][10000::]
        dict[label[i]]['HBond_int']['ns'] = HBond[1][10000::]
    else:
        dict[label[i]]['HBond_int']['ts'] = HBond[0][10000::]
        dict[label[i]]['HBond_int']['ns'] = HBond[1][10000::]
print(len(dict['20gly']['HBond_int']['ns']))
print(len(dict['alt']['HBond_int']['ns']))

for i in range(len(filelist)):  
    plt.subplot(2,4,i+1)
    plt.title(label[i])
    plt.hist(dict[label[i]]['HBond_int']['ns'], bins = 11, range = [-.5, 10.5], color = colour[i], 
             label = label[i], histtype = 'step')
    plt.xlabel('# of HBonds')
    plt.ylabel('Counts')
    plt.ylim(0, 15500)

plt.tight_layout(rect = [0, 0, 2, 1.])
plt.savefig('./plots/intraHBond.pdf', dpi = 300, bbox_inches = 'tight')

```

# Ree vs. Rg

```python
### End-to-end distance and radius of gyration ###

n = 1

filelist = ['dist-ree-20gly.xvg', 'dist-ree-alt.xvg', 'dist-ree-ran.xvg', 'dist-ree-dib.xvg', 
            'dist-ree-2ran.xvg', 'dist-ree-A.xvg', 'dist-ree-B.xvg', 'dist-ree-C.xvg']

for i in range(len(filelist)):
    dict[label[i]]['dist-ree'] = {}
    res = np.loadtxt(filelist[i], comments = ['#', '@'], unpack = 'true')
    
    if label[i] == '20gly':
        dict[label[i]]['dist-ree']['r'] = res[1][10000::] #### HERE THE SPACING MATTERS
    else:
        dict[label[i]]['dist-ree']['r'] = res[1][10000::]
        
        
    # Here goes the inserted python script!
    x_min, x_max, y_min, y_max = 0., 5., 0.2, 4.6

    # Number of 2D regions in which the plot is divided.
    x_resolution, y_resolution = 240, 240 #120, 120

    #points = [read_angles(line) for line in open(data_file)]
    print(np.min(dict[label[i]]['rg']['r']))
    
    points = np.zeros((len(dict[label[i]]['rg']['r']), 2))
    for j in range(len(dict[label[i]]['rg']['r'])):
        points[j][0] = dict[label[i]]['rg']['r'][j]
        points[j][1] = dict[label[i]]['dist-ree']['r'][j]
        
    #points = np.transpose([dict[label[i]]['rg']['r'], dict[label[i]]['dist-ree']['r']])
    count = len(points)
    print(points)
    histogram = np.zeros([x_resolution, y_resolution])
    x_interval_length = (x_max - x_min) / x_resolution
    y_interval_length = (y_max - y_min) / y_resolution
    interval_surface = x_interval_length * y_interval_length
    increment = 1000.0 / count / interval_surface

    for item in points:
        x = int((item[0] - x_min) / x_interval_length)
        y = int((item[1] - y_min) / y_interval_length)
        histogram[x,y] += increment

    x_intervals = np.arange(x_min, x_max, (x_max - x_min) / x_resolution)
    y_intervals = np.arange(y_min, y_max, (y_max - y_min) / y_resolution)
    
    
    histogram_file = 'hist.txt'
    o = open(histogram_file, 'w')
    for i, x in enumerate(x_intervals):
        for j, y in enumerate(y_intervals):
             o.write('%f %f %f \n' % (x, y, histogram[i, j]))
        #o.write('\n')
        
    with open("hist.txt", 'r') as r, open('hist_new.txt', 'w') as o:
        for line in r:
            #strip() function
            if line.strip():
                o.write(line)
    f = open("hist.txt", "r")

    print(histogram.max())
    z = np.reshape(np.loadtxt('hist_new.txt', unpack = True)[2], (x_resolution, y_resolution))
    
    print(n)
    plt.subplot(2, 4, n)
    plt.title(label[n-1])
    plt.contourf(x_intervals, y_intervals, z)
    plt.xlim(.3, 2.5)
    plt.ylim(.5, .9)
    plt.ylabel('$R_g$ [nm]')
    plt.xlabel('$R_{ee}$ [nm]')
    n += 1
    
plt.tight_layout(rect = [0, 0, 2, 1.])
plt.savefig('./plots/RgRee.pdf', dpi = 300, bbox_inches = 'tight')
```
